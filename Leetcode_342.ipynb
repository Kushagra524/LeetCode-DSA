***Best Solution***

```python
class Solutions:
  def isPowerOfFour(self , n : int) -> bool:
    while n>0 and n % 4 == 0:    # jab tak n > 0 and n 4 se divisible hai , tab tak n ko 4 se divide kro
      n //= 4
    return n == 1    # if last me 1 aya then True otherwise False

ss = Solutions()

print(ss.isPowerOfFour(26))
print(ss.isPowerOfFour(128))
print(ss.isPowerOfFour(16))
```

```python
import math
class Solution:
  def isPowerOfFour(self , n : int) -> bool:
    if n <= 0:
      return False

    else:
      return math.log(n , 4).is_integer()

s = Solution()
```

```python
print(s.isPowerOfFour(64))
print(s.isPowerOfFour(14))
print(s.isPowerOfFour(10))
```

***Using Recursion***

```python
class Solution2:
  def isPowerOfFour(self , n : int) -> bool:
    # making my base case
    if n <=0:
      return False
    if n == 1:
      return True
    if n %4 != 0:
      return False

    # makin my recursive case
    return self.isPowerOfFour(n // 4)

s2 = Solution2()
```

```python
print(s2.isPowerOfFour(64))
print(s2.isPowerOfFour(14))
print(s2.isPowerOfFour(10))
```

***Using WHile Loop***

```python
class Solution3:
  def isPowerOfFour(self , n : int) -> bool:
    while n % 4 == 0:
      n = n // 4

    if n == 1:
      return True
    else:
      return False

s3 = Solution3()
```

```python
print(s3.isPowerOfFour(64))
print(s3.isPowerOfFour(14))
print(s3.isPowerOfFour(10))
```
