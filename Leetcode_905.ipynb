# ***905. Sort Array by Parity***

image.png

```python
# example array = [5 , 1 , 6 , 2 , 7 , 4 , 3]

# answer = [6 , 2 , 4 , 5 , 1 , 7 , 3] or [2 , 6 , 4 , 3 , 1 , 7 , 5] or any combination but even elements must be ar starting.
```

```python
# so what we do is , we take a start index as reference and iterate from index = 1 to n , if i is even we just swap it by start and increase start variable by 1 , if i odd , nothing happens.
```

```python
from typing import List

class Solution:
  def sortArrayByParity(self , nums : List[int]) -> List[int]:
    n = len(nums)

    if n <= 1:
      return nums

    start = 0
    for i in range(0 , n):
      if nums[i] % 2 == 0:
        nums[start] , nums[i] = nums[i] , nums[start]
        start = start + 1
    return nums

s = Solution()
```

```python
print(s.sortArrayByParity([5 , 1 , 6 , 2 , 7 , 4 , 3]))
```

```python
print(s.sortArrayByParity([3 , 1 , 2 , 4]))
```

```python
print(s.sortArrayByParity([1 , 0]))
```

```python
print(s.sortArrayByParity([0 , 1 , 2]))
```
