***My first approach is to sort my strings and then compare if they r equal ot not***

```python
class Solution:
  def validAnagram(self , s : str , t : str) -> bool:
    s = "".join(sorted(s))
    t = "".join(sorted(t))

    if s == t:
      return True

    else:
      return False

s = Solution()
```

```python
print(s.validAnagram(s = 'ggii' , t = 'eekk'))
```

***Another approach using hash_map***

```python
import collections

class Solution:
  def isAnagram(self , s : str , t : str) -> bool:
    map_s = collections.Counter(s)
    map_t = collections.Counter(t)

    if map_s == map_t:
      return True
    else:
      return False

s = Solution()
```

```python
print(s.isAnagram(s = "ggkk" , t = "ggkkk"))
```

***Optimised approach of hash_map***

```python
import collections

class Solutions:
  def anagram(self , s : str , t : str) -> bool:
    if len(s) != len(t):
      return False
    s = s.lower()
    t = t.lower()
    map_s = collections.Counter(s)

    for char in t:
      map_s[char] = map_s[char] - 1
      if map_s[char] < 0:
        return False

    for counter in map_s.values():
      if counter != 0:
        return False # If any count is not zero, it's not an anagram

    return True # If the loop completes, all counts were zero, so it is an anagram

s1 = Solutions()
```

```python
print(s1.anagram(s = 'Kirti' , t = 'Kirti'))
```
