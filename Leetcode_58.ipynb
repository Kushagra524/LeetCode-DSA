# ***58. Length of Last Word***

image.png

```python
class Solution:
  def lastWord(self , s : str) -> int:
    s = s.strip()    # .strip() removes the traling zeros from the string

    i = -1
    while (i >= -(len(s))) and (s[i] != " "):
      i = i - 1

    i = -1*(i + 1)
    return i

s = Solution()
```

*Explanation : I will iterate from last and basically I will see that the first moment that I get a empty space i will stop .*

*Then that particular index will be added with 1 becoz , while loop iterates through 1 extra , and then at last I will convert i into a +ve number and return it .*

```python
print(s.lastWord("Hello World"))
print(s.lastWord("   fly me   to   the moon  "))
```
