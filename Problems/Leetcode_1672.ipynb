# ***1672. Richest Customer Wealth***

image.png

```python
from typing import List

class Solution:
  def maxWealth(self , accounts : List[List[int]]) -> int:
    max_wealth = 0
    for acc in accounts:
      max_wealth = max(max_wealth , sum(acc))
    return max_wealth
```

```python
s = Solution()
s.maxWealth([[1,2,3],[3,2,1]])
```

```python
s.maxWealth([[1,5],[7,3],[3,5]])
```
