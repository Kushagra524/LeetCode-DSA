# ***1431. Kids With the Greatest Number of Candies***

image.png

```python
from typing import List
```

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        result = []
        count = max(candies)
        for i in candies:
            if (i + extraCandies) >= count:
                result.append(True)
            else:
                result.append(False)
        return result
```

```python
s = Solution()
```

```python
s.kidsWithCandies(candies = [2,3,3,5,4,6] , extraCandies = 3)
```
