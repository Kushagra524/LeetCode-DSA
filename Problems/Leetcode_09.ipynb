# ***Palindrome Number***

:Screenshot 2026-06-23 220552.png

```python
class Solution:
  def isPalindrome(self , x : int) -> bool:
    temp = x
    result = 0
    while x > 0:
      digit = x % 10
      result = result * 10 + digit
      x = x // 10

    if temp == result or result == temp:
      return True
    else:
      return False
```

```python
s = Solution()
s.isPalindrome(121)
```
