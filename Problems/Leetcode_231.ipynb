# ***Power of Two***

image.png

```python
import math
class Solution:
  def isPowerOfTwo(self , n : int) -> bool:
    if n <= 0:
      return False
    else:
      return math.log2(n).is_integer()

s = Solution()
```

```python
print(s.isPowerOfTwo(25))
print(s.isPowerOfTwo(16))
print(s.isPowerOfTwo(36))
```

***Another Mehtod using while Loop***

```python
class Solution2:
  def isPowerOfTwo(self , n : int) -> bool:
    while n % 2 == 0:
      n = n // 2

    if n == 1:
      return True
    else:
      return False
s2 = Solution2()
```

```python
print(s2.isPowerOfTwo(25))
print(s2.isPowerOfTwo(64))
```

***Another method using Recursion***

```python
class Solution3:
  def isPowerOfTwo(self , n : int) -> bool:
    # base case will be
    if n <= 0:
      return False
    if n == 1:
      return True
    if n % 2 == 1:
      return False

    # recursive case
    return self.isPowerOfTwo(n//2)

s3 = Solution3()
```

```python
print(s3.isPowerOfTwo(100))
print(s3.isPowerOfTwo(128))
```
