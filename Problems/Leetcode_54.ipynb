First I decide the starting and ending of my columns.

And what i do is itereate through col_start = 0 till col_end + 1 .

then what I see is I have already covered row = 0 and append those elements into my result = [].

now row_start += 1

now we print col_end from row_start till row_end.
and append its elements into result.

col_end = col_end - 1



```python
from typing import List

class Solution:
  def spiralOrder(self , matrix : List[List[int]]) -> List[int]:
    n = len(matrix)
    m = len(matrix[0])

    col_start = 0
    col_end = m - 1

    row_start = 0
    row_end = n - 1

    total = n * m
    result = []
    count = 0

    while count < total:

      for i in range(col_start , col_end + 1):
        result.append(matrix[row_start][i])
        count = count + 1
      row_start = row_start + 1
      if count ==  total:
        break

      for i in range(row_start , row_end + 1):
        result.append(matrix[i][col_end])
        count = count + 1
      col_end = col_end - 1
      if count == total:
        break

      for i in range(col_end , col_start - 1 , -1):
        result.append(matrix[row_end][i])
        count = count + 1
      row_end = row_end - 1
      if count ==  total:
        break

      for i in range(row_end , row_start - 1 , -1):
        result.append(matrix[i][col_start])
        count = count + 1
      col_start = col_start + 1
      if count ==  total:
        break

    return result


s = Solution()
```

```python
print(s.spiralOrder([[1,2,3],[4,5,6],[7,8,9]]))
```
