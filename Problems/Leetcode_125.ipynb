```python
class Solution:
  def isPalindrome(self , s : str) -> bool:
    word = s.lower()

    # first I will remove the alpha numeric characters
    word = ''.join(character for character in word if character.isalnum())

    # now I will use two pointers logic
    left = 0
    right = len(word) - 1

    while left < right:
      if word[left] != word[right]:
        return False

      left += 1
      right -= 1
    return True


s = Solution()
```

```python
print(s.isPalindrome("A man, a plan, a canal: Panama"))
```

***Another approach***

```python
class Solutions:
  def isalpha(self , s):
    ascii = ord(s)
    if (ascii >= 65 and ascii <= 90) or (ascii >= 97 and ascii <= 122) or (ascii >= 48 and ascii <= 57):
      return True
    return False

  def isPalindrome(self , s : str) -> bool:
    s = s.lower()
    left = 0
    right = len(s) - 1

    while left < right:
      if not self.isalpha(s[left]):
        left += 1
      elif not self.isalpha(s[right]):
        right -= 1
      elif s[left] == s[right]:
        left += 1
        right -= 1
      else:
        return False
    return True

s1 = Solutions()
```

```python
print(s1.isPalindrome("A man, a plan, a canal: Panama"))
```
