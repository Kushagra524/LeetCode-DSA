```python
from typing import List

class Solution:
  def addToArrayForm(self , num : List[int] , k : int) -> List[int]:
    i = len(num) - 1   # initially i = 3 , for example of num = [1 , 2 , 0 , 0]
    carry = k          # k = 34
    result = []        # empty list

    while i >= 0 or carry > 0:
      if i >= 0:
        carry = carry + num[i]
        i = i - 1

      result.append(carry % 10)
      carry = carry // 10

    return result[ : : -1]

s = Solution()
```

```python
print(s.addToArrayForm([1,2,0,0] , k = 34))
```

```python
print(s.addToArrayForm([2 , 1 , 5] , k = 816))
```
