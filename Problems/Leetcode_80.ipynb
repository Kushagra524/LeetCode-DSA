```python
# question says let an array a = [ 3 , 3 , 3 , 7 , 9 , 9 , 9 , 9 , 12 , 12]
# result = [3 , 3 , 7 , 9 , 9 , 12 , 12 , 3 , 9 , 9 , 9]

# now we have to put the unique elements in sorted form but but but , the element can appear atmost twice. as we can see the example above
```

```python
from typing import List

class Solution:
  def removeDuplicates(self , nums : List[int]) -> List:
    start = 1

    for i in range(2 , len(nums)):
      if nums[i] != nums[start - 1]:
        start = start + 1
        nums[start] = nums[i]
    return start + 1

s = Solution()
```

```python
print(s.removeDuplicates([3 , 3 , 3 , 7 , 9 , 9 , 9 , 9 , 12 , 12]))
```

*Explanation : I took starting index (start = 1) becoz , first 2 elements maybe same or different doesn't matters , as it is already given in the question that unique element can occur atmost twice.*
