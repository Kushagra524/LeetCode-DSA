# ***First approach - Two Pointers***

```python
from typing import List

class Solution:
  def twoSum(self , numbers : List[int] , target : int) -> List[int]:
    # here I will use two pointer approach
    left = 0
    right = len(numbers) - 1

    while left < right:
      sum = numbers[left] + numbers[right]
      if sum == target:
        return [left + 1 , right + 1]

      elif sum > target:
        right = right - 1

      elif sum < target:
        left = left + 1

s = Solution()
```

```python
print(s.twoSum(numbers = [3 , 78 , 91 , 103] , target = 106))
```

# ***Second approach - Hash Map***

```python
class Solutions:
  def twoSumSorted(self , nums : List[int] , target : int) -> List[int]:
    hash_map = {}

    for i in range(len(nums)):
      diff = target - nums[i]
      if diff == 0:
        return [i + 1]       # I did + 1 becoz it is a 1-indexed array

      if diff in hash_map:
        return [hash_map[diff] + 1 , i + 1]   # here also + 1

      else:
        hash_map[nums[i]] = i

s1 = Solutions()
```

```python
print(s1.twoSumSorted(nums = [23 , 54 , 71 , 62] , target = 23))
```
