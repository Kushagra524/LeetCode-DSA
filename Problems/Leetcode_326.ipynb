***Best solution to outbeat Runtime than others***

```python
class SOlution:
  def isPowerOfThree(self , n : int) -> bool:
    while n > 0 and n % 3 == 0:
      n = n // 3
    return n == 1

s = SOlution()
```

```python
print(s.isPowerOfThree(27))
print(s.isPowerOfThree(15))
print(s.isPowerOfThree(36))
```
