```python
# acc. to question : example arr = [3 , 3 , 5 , 7 , 7 , 7 , 9 , 9 , 9 , 12 , 12]

# result array = [3 , 5 , 7 ,9 , 12 , , ,  ....]
# num_of_unique_elements = value calculates = 5

# T.C = O(n) & S.C = O(1)

# now if and element if equal to it's previous element then it is not a unique element , otherwise unique element.

# now first element is always unique becoz no previous elemnt equal to it.
```

```python
from typing import List

class Solution:
  def removeDuplicates(self , nums : List[int]) -> int:
    if not nums:
      return 0
    start = 0  # this is our first element (index = 0)
    for i in range(1 , len(nums)):    # we are iterating from 1 to n

      # this is my condition for unique element in my array
      if nums[i] != nums[start]:
        start += 1  # in case of unique value = start increases
        nums[start] = nums[i]  # and the unique element is shifted to the index of num[start]
    return start + 1    # number of unique elements I get = start + 1
s = Solution()
```

```python
print(s.removeDuplicates([3 , 3 , 5 , 7 , 7 , 7 , 9 , 9 , 9 , 12 , 12]))
print(s.removeDuplicates([5 , 5 , 6 , 7 ,8 , 8 , 8 , 9 , 10 , 11 , 11 , 11 , 12]))
```
