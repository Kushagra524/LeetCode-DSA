```python
class Solution:
  def firstUniqueChar(self , s : str) -> int:
    char_counts = {}


    for char in s:
      char_counts[char] = char_counts.get(char, 0) + 1

    for i in range(len(s)):
      if char_counts[s[i]] == 1:
        return i

    else:
      return -1
```

```python
s = Solution()
```

```python
print(s.firstUniqueChar('leetcode'))
print(s.firstUniqueChar('loveleetcode'))
print(s.firstUniqueChar('new_mansion'))
```

```python
import collections


class Solutions:
  def firstUniqChar(self , s : str) -> int:
    char_counts = collections.Counter(s)     # .enumerate() is a built-in funciton in Python that is used to store the item and its frequency into a dictionary.

    for index , char in enumerate(s):    # i is the index and char is its value
      if char_counts[char] == 1:
        return index

    else:
      return -1

s1= Solutions()
```

```python
print(s1.firstUniqChar('leetcode'))
print(s1.firstUniqChar('loveleetcode'))
print(s1.firstUniqChar('new_mansion'))
```
