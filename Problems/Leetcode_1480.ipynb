# ***Prefix Sum / Running Sum***

image.png

```python
# running sum also means prefix sum = means sum from index 0 upto a a specific index ===> like example array = [3 , 1 , 8 , 2 , 10 , 5]

# running / prefix sum = [3 , 4 , 12 , 14 , 24  , 29]
```

```python
from typing import List

class Solution:
  def runningSum(self , nums : List[int]) -> List[int]:
    n = len(nums)

    result = []
    result.append(nums[0])

    for i in range(1 , n):
      result.append(result[i-1] + nums[i])

    return result

s = Solution()
print(s.runningSum([3 , 1 , 8 , 2 , 10 , 5]))
print(s.runningSum([1 , 2 , 3 , 4]))
print(s.runningSum([1 , 1 , 1 , 1 , 1]))
```

***Another method - which uses same approach but less space than above code.(called in-place solution that says , no need to make a new array / list instead modify the existing one)***

```python
from typing import List
class Solutions:
  def runningSum(self , nums : List[int]) -> List[int]:

    for i in range(1  ,len(nums)):   # instead of making a new list , i modified the given list only
      nums[i] += nums[i - 1]   # but what about nums[0]  ====> nums[0] remains same as it is , becoz i never changed

    return nums

s = Solutions()
```

```python
print(s.runningSum([3 , 1 , 8 , 2 , 10 , 5]))
print(s.runningSum([1 , 2 , 3 , 4]))
```
