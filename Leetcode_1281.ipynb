# ***1281. Subtract Product and Sum of Digits of an Integer***

image.png

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        sum = 0
        product = 1

        temp = n
        while temp > 0:
            id = temp % 10
            sum += id
            product *= id
            temp = temp // 10
        return product - sum
```

```python
s = Solution()
```

```python
s.subtractProductAndSum(121)
```

```python
s.subtractProductAndSum(1248)
```
