```python
from typing import List

class Solution:
  def reverseString(self , s : List[str]) -> None:   # None means this function doesn't returns anything but for checking my output , I returned the string
    s.reverse()
    return s

s = Solution()
```

```python
print(s.reverseString(['k' , 'u' , 's' , 'h']))
```

***Better Approach***

```python
class Solutions:
  def reversedString(self , s : List[str]) -> None:
    # here I will apply 2 pointer approach

    left = 0
    right = len(s) - 1

    while left < right:
      s[left] , s[right] = s[right] , s[left]
      left = left + 1
      right = right - 1
    return s

s1 = Solutions()
```

```python
print(s1.reversedString(['k' , 'u' , 's' , 'h']))
```
