# ***66. Plus One***

image.png

```python
from typing import List

class Solution:
  def plusOne(self, digits: List[int]) -> List[int]:
    num = 0
    for i in digits:
      num = num * 10 + i   # first I extract the numbers from digits list

    num = num + 1
    s = str(num)
    digits = []
    for i in s:
      digits.append(int(i))

    return digits

s = Solution()
```

```python
print(s.plusOne([9]))
```
