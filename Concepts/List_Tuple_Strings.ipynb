# ***Tuple***

*A tuple is a built-in data type in Python that is used to store multiple items in a single variable. It is similar to a list, but the main difference is that a tuple is immutable, which means its elements cannot be changed, added, or removed after it is created. Tuples are created using parentheses () and can store different types of data, such as numbers, strings, and Boolean values. They are useful when you want to keep data fixed and protected from accidental changes.*

Properties of a Tuple in Python

Ordered – Elements are stored in a fixed order and can be accessed using their index.

Immutable – Once a tuple is created, its elements cannot be added, removed, or modified.

Allows Duplicate Values – A tuple can contain the same value multiple times.

Can Store Different Data Types – A tuple can hold integers, strings, floats, booleans, and even other tuples.

Indexed – Each element has an index starting from 0, making it easy to access specific values.

Supports Slicing – You can retrieve a portion of a tuple using slicing (tuple[start:end]).

Can Be Nested – A tuple can contain another tuple as one of its elements.

Faster Than Lists – Tuples generally use less memory and are slightly faster than lists because they are immutable.

Can Be Used as Dictionary Keys – Since tuples are immutable, they can be used as keys in dictionaries (if all their elements are hashable).

Supports Iteration – You can loop through the elements of a tuple using for or while loops.

```python
# Python Tuple Methods and Operations


# 1. Creating a Tuple
names = ("Kushagra", "Kirti", "Rahul", "Ananya", "Rohan")

print("Original Tuple:")
print(names)
print("-> A tuple stores multiple values in a single variable.\n")


# 2. len()
print("Length of Tuple:", len(names))
print("-> Returns the total number of elements.\n")


# 3. count()
print("Count of 'Kirti':", names.count("Kirti"))
print("-> Counts how many times an element appears.\n")


# 4. index()
print("Index of 'Rahul':", names.index("Rahul"))
print("-> Returns the index of the given element.\n")


# 5. in
print("'Kirti' in tuple:", "Kirti" in names)
print("-> Checks whether an element exists.\n")


# 6. not in
print("'Aman' not in tuple:", "Aman" not in names)
print("-> Checks whether an element does not exist.\n")


# 7. min()
print("Minimum (Alphabetically):", min(names))
print("-> Returns the alphabetically smallest element.\n")


# 8. max()
print("Maximum (Alphabetically):", max(names))
print("-> Returns the alphabetically largest element.\n")


# 9. Tuple Slicing
print("First 3 Names:", names[:3])
print("Last 2 Names:", names[-2:])
print("-> Extracts a portion of the tuple.\n")


# 10. Tuple Concatenation
extra_names = ("Sakshi", "Aditya")
combined = names + extra_names

print("Combined Tuple:")
print(combined)
print("-> Joins two tuples into a new tuple.\n")


# 11. Tuple Repetition
print("Repeated Tuple:")
print(("Hello",) * 3)
print("-> Repeats the tuple multiple times.\n")


# 12. Nested Tuple
student = ("Kushagra", 20, ("Python", "Machine Learning"))

print("Nested Tuple:")
print(student)
print("-> A tuple can contain another tuple.\n")


# 13. Iterating through a Tuple
print("Printing all elements using a loop:")
for name in names:
    print(name)

print("-> Tuples can be traversed using loops.\n")


# 14. Type Conversion
temp_list = list(names)

print("Tuple converted to List:")
print(temp_list)
print("-> Tuples can be converted into lists if modification is needed.\n")


print("🎉 Successfully demonstrated all common Python tuple methods and operations!")
```

# ***String***

*A string is a sequence of characters used to store and represent text in Python. It can contain letters, numbers, symbols, and spaces, and is enclosed in single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """). Strings are immutable, which means their characters cannot be changed after they are created. Python provides many built-in methods to perform operations such as searching, replacing, splitting, joining, and changing the case of strings, making them very useful for handling textual data.*

Properties of Strings

Ordered – Characters are stored in a fixed sequence.

Immutable – Characters cannot be modified after the string is created.

Indexed – Each character has an index starting from 0.

Supports Negative Indexing – Characters can also be accessed from the end using negative indexes (-1, -2, etc.).

Supports Slicing – A portion of a string can be extracted using slicing (string[start:end]).

Allows Duplicate Characters – The same character can appear multiple times in a string.

Can Store Letters, Numbers, Symbols, and Spaces – A string can contain any valid character.

Supports Concatenation – Two or more strings can be joined using the + operator.

Supports Repetition – A string can be repeated multiple times using the * operator.

Iterable – Each character can be accessed using loops like for or while.

Rich Built-in Methods – Strings provide many useful methods such as upper(), lower(), replace(), split(), strip(), find(), and count() for text manipulation.

```python
text = 'Kushagra is a RAG and Machine Learning Specialist.'

# find()
print("Index of 'Python':", text.find("Python"))

# index()
print("Index of 'RAG':", text.index("RAG"))

# replace()
print("After Replace:", text.replace("Machine", "Deep"))

# startswith()
print("Starts with 'Kushagra':", text.startswith("Kushagra"))

# endswith()
print("Ends with 'Machine':", text.endswith("Machine"))

# split()
words = text.split()

# in
print("'Machine' in string:", "Machine" in text)

# not in
print("'Machine' not in string:", "Machine" not in text)

# Slicing
print("First 8 Characters:", text[:8])
print("Last 11 Characters:", text[-11:])

# Concatenation
new_text = text + " is amazing."
print(new_text)
```

```python
# Remaining Python String Methods

text = "Kushagra loves Python Programming"


# casefold()
print("Casefold:", text.casefold())
print("-> Converts the string to lowercase (stronger than lower()).\n")


# encode()
print("Encoded:", text.encode())
print("-> Converts the string into bytes.\n")


# expandtabs()
tab_text = "Python\tProgramming"
print("After expandtabs():")
print(tab_text.expandtabs(15))
print("-> Replaces tab spaces with the specified number of spaces.\n")


# format()
name = "Kushagra"
language = "Python"
print("Formatted String:")
print("{} loves {}.".format(name, language))
print("-> Inserts values into placeholders.\n")


# format_map()
data = {"name": "Kushagra", "course": "Python"}
print("Format Map:")
print("{name} is learning {course}.".format_map(data))
print("-> Inserts values from a dictionary.\n")


# isascii()
print("Is ASCII:", text.isascii())
print("-> Checks whether all characters are ASCII.\n")


# isdecimal()
print("'12345' is decimal:", "12345".isdecimal())
print("-> Checks whether all characters are decimal digits.\n")


# isnumeric()
print("'12345' is numeric:", "12345".isnumeric())
print("-> Checks whether all characters are numeric.\n")


# isidentifier()
print("'student_name' is identifier:", "student_name".isidentifier())
print("-> Checks whether the string is a valid Python identifier.\n")


# isprintable()
print("Is Printable:", text.isprintable())
print("-> Checks whether all characters are printable.\n")


# isspace()
print("'   ' is space:", "   ".isspace())
print("-> Checks whether the string contains only whitespace.\n")


# istitle()
print("'Python Programming' is title:", "Python Programming".istitle())
print("-> Checks whether each word starts with a capital letter.\n")


# ljust()
print("Left Justified:")
print(text.ljust(45, "-"))
print("-> Aligns the string to the left.\n")


# rjust()
print("Right Justified:")
print(text.rjust(45, "-"))
print("-> Aligns the string to the right.\n")


# zfill()
number = "25"
print("Zero Filled:", number.zfill(5))
print("-> Adds leading zeros to the string.\n")


# partition()
print("Partition:")
print(text.partition("Python"))
print("-> Splits the string into three parts.\n")


# rpartition()
print("Reverse Partition:")
print(text.rpartition("Python"))
print("-> Splits from the last occurrence.\n")


# splitlines()
lines = "Python\nJava\nC++"
print("Split Lines:")
print(lines.splitlines())
print("-> Splits the string into separate lines.\n")


# rfind()
print("Reverse Find:", text.rfind("o"))
print("-> Returns the last occurrence of a character.\n")


# rindex()
print("Reverse Index:", text.rindex("o"))
print("-> Returns the last index of a character.\n")
```

***Leetcode - 1480. Running Sum of 1D Array***

image.png

```python
# running sum also means prefix sum = means sum from index 0 upto a a specific index ===> like example array = [3 , 1 , 8 , 2 , 10 , 5]

# running / prefix sum = [3 , 4 , 12 , 14 , 24  , 29]
```

```python
from typing import List

class Solution:
  def runningSum(self , nums : List[int]) -> List[int]:
    n = len(nums)
    sum = []
    sum.append(nums[0])

    for i in range(1 , n):
      sum.append(sum[i - 1] + nums[i])

    return sum

s = Solution()
```

```python
print(s.runningSum([3 , 1 , 8 , 2 , 10 , 5]))
print(s.runningSum([1 , 2 , 3 , 4]))
```

***Another optimized approach is : in-place Solution (which means , no need to make a new array / list , instead modify the existing one)***

```python
class Solutions:
  def runningSum(self , nums : List[int]) -> List[int]:

    for i in range(1 , len(nums)):
      nums[i] = nums[i] + nums[i - 1]    # instead of creating another list , i will modify the same list.

      # at i = 1 nums = [num[1] + num[0]].
      # at i = 2 , nums = [num[1] + num[0] , num[2] + num[1] + num[0]]
      # and so on.

      # but what about num[0] = it it already present , becoz we never modified it.

    return nums

s = Solutions()
```

```python
print(s.runningSum([3 , 1 , 8 , 2 , 10 , 5]))
print(s.runningSum([1 , 2 , 3 , 4]))
```
