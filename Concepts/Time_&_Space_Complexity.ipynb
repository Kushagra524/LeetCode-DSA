# ***Time Complexity***

#### *Describes how the execution time of an algo. grows as the input size increses. It essentially quantifies the number of operations an algorithm performs as a function of the input data size.*

```python
# example - 1

for i in range(n):
  print(i)

# time complexity - O(n)
```

***Linear Time Complexity = O(n)***

```python
# example - 1
for i in range(n):
  print(i , end = ', ')   # O(n)

# example - 2
for i in range(n + 100):
  print(i , end= ', ')    # O(n)

# example - 3
for i in range(k * n):
  print(i , end = ', ')   # O(n)
```

***Constant Time Complexity = O(1)***

```python
# example - 1
for i in range(100000):
  print(i , end =', ')     # O(1)

# example - 2
for i in range(5999999):
  print(i , end =', ')     # O(1)
```

***Quadratic Time Complexity = O(n²)***

```python
# example - 1
n = 10 or 100 or anything
for i in range(n**2):
  print(i , end =', ')    # O(n²)

# example - 2
for i in range(n):
  for j in range(n):
    print(i , end =', ')    # O(n²)
```

***Logarithmic Time Complexity = O(logn)***

```python
# example - 1
n = 100
while n > 0:
  print(n , end = ' ')    # O(logn)
  n = n // 2

  # here output is 100 then 50 then 25 then 12 then 6 then 3 then 1 and so on.
  # this is logarithmix time complxity
```

```python
# example - 2
i = 1
n = 100

while i < n:
  print(i , end = ' ')
  i = i * 2

# explanation - suppose k operations are performed , so output comes as 2^0 , 2^1 , 2^2  ,2^3 , ........... , 2^k

# for worst case : we take 2 ^ k = 100  ===> taking log both sides ====> log( 2^k ) = log(100)

# ==> k.log (2) = log(100)
# ==> k = log(100) / log(2)  ===> k = log2(100)  or log (100) base 2
# and for n = n time complexity = O(logn)
```

***Linearithmic / Log-linear = O(nlogn)***

```python
for i in range(n):          # n times
  while i < n:
    print(i , end = ' ')    # logn times
    i = i * 2               # total = O(nlogn)
```

***Comparison of Time Complexities : O(1) < O(log n) < O(n) < O(n log n) < O(n²) < O(n³) < O(2ⁿ) < O(n!)***
