# ***Set*** -

collections of unique data , meaning that elements within a set cannot be duplicated.

```python
print('1. Set Definition: Creating a set from a list of elements.')
my_set = {1, 2, 3, 4, 5}
print(f'Original set: {my_set}')
```

```python
print('\n2. Adding Elements: Using .add() to add a single element.')
my_set.add(6)
print(f'After adding 6: {my_set}')
```

```python
print('\n3. Removing Elements: Using .remove() (raises error if not present) and .discard() (no error if not present).')

my_set.remove(3)   # raises error if not present in set

print(f'After removing 3 with .remove(): {my_set}')

my_set.discard(10) # 10 is not in the set, no error

print(f'After discarding 10: {my_set}')
```

```python
print('\n4. Union: Combining elements from two sets (no duplicates).')

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

union_set = set_a.union(set_b)

print(f'Set A: {set_a}, Set B: {set_b}')
print(f'Union (A | B): {union_set}')
```

```python
print('\n5. Intersection: Elements common to both sets.')

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

intersection_set = set_a.intersection(set_b)

print(f'Set A: {set_a}, Set B: {set_b}')
print(f'Intersection (A & B): {intersection_set}')
```

```python
print('\n6. Difference: Elements in the first set but not in the second.')

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

difference_set_ab = set_a.difference(set_b)
difference_set_ba = set_b - set_a

print(f'Set A: {set_a}, Set B: {set_b}')
print(f'Difference (A - B): {difference_set_ab}')
print(f'Difference (B - A): {difference_set_ba}')
```

```python
print('\n7. Symmetric Difference: Elements in either set, but not in both.')

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

symmetric_difference_set = set_a.symmetric_difference(set_b)

print(f'Set A: {set_a}, Set B: {set_b}')
print(f'Symmetric Difference (A ^ B): {symmetric_difference_set}')
```

```python
print('\n8. Subset and Superset: Checking relationships between sets.')

set_x = {1, 2, 3}
set_y = {1, 2, 3, 4, 5}

print(f'Set X: {set_x}, Set Y: {set_y}')

print(f'Is X a subset of Y? {set_x.issubset(set_y)}')
print(f'Is Y a superset of X? {set_y.issuperset(set_x)}')
print(f'Is Y a subset of X? {set_y.issubset(set_x)}') # False
```

```python
print('\n9. Clearing a Set: Removing all elements from a set.')

my_set_to_clear = {10, 20, 30}

print(f'Set before clearing: {my_set_to_clear}')
my_set_to_clear.clear()
print(f'Set after clearing: {my_set_to_clear}')
```

```python
print('\n10. Disjoint Sets: Checking if two sets have no common elements.')

set_c = {1, 2, 3}
set_d = {4, 5, 6}
set_e = {3, 7, 8}

print(f'Set C: {set_c}, Set D: {set_d}, Set E: {set_e}')

print(f'Are C and D disjoint? {set_c.isdisjoint(set_d)}')
print(f'Are C and E disjoint? {set_c.isdisjoint(set_e)}')
```

```python
print('\n11. Intersection Update: Retaining only common elements.')

set_f = {1, 2, 3, 4, 5}
set_g = {4, 5, 6, 7, 8}

print(f'Set F before update: {set_f}')
print(f'Set G: {set_g}')

set_f.intersection_update(set_g)

print(f'Set F after intersection_update with G: {set_f}')
```

```python
print('\n12. Difference Update: Removing elements found in another set.')

set_h = {10, 20, 30, 40, 50}
set_i = {30, 40, 60, 70}

print(f'Set H before update: {set_h}')
print(f'Set I: {set_i}')

set_h.difference_update(set_i)

print(f'Set H after difference_update with I: {set_h}')
```

```python
print('\n13. Checking presence of any element.')

set_j = {'Kushagra' , 'Ayush' , 'Kirti'}

if 'Kirti' in set_j:
  print('Yess')
```

```python
print('\n14. Some more concepts about set are : ')

set_k = {1 , 2 , 3 , 4}
set_l = set_k   # I copied set_k into set_l

set_l.add(5)    # now if I make changes in set_l , it will also changes int set_k , becoz not only the values but the reference of both sets has now become same.

print(f"CHanges made into set_l(elements = 5 is added) : {set_l}")
print(f"Previous set also updates (with same operation on it i.e. elemnts = 5 added): {set_k}")
```

```python
print('\n15. But if i only wanna copy values into another set.')

set_1 = {1 , 2 , 3 , 4}

set_2 = set_1.copy()   # I wil use .copy() function to copy only the values not the reference
set_2.add(5)          # Now if I wanna perform operations on new set , it won't affect the previous set

print(f"UPdated new set with operation peformed(elements = 5 added) : {set_2}")
print(f"Older set didn't had any changes : {set_1}")
```

# ***Dictionary*** -

Collection of items , similar to lists and tuples. Unlike lists and tuples , each item in a dictionary is a key_value pair {key : value}.

```python
print('1. Dictionary Definition: Creating a dictionary.')

my_dict = {
    'name': 'Kushagra',
    'age': 22,
    'city': 'Bhopal'
}

print(f'Original dictionary: {my_dict}')
```

```python
print('\n2. Accessing Elements: Using keys to retrieve values.')

print(f"Name: {my_dict['name']}")
print(f"Age: {my_dict['age']}")

# Using .get() method for safe access (returns None if key not found)
print(f"City using .get() method : {my_dict.get('city')}")
print(f"Country using .get() method with default: {my_dict.get('country', 'Unknown')}")
print(f"Retrieving a key value that isn't present in the dictionary (gives me None) : {my_dict.get('salary')}")
```

```python
print('\n3. Adding and Updating Elements: Assigning values to keys.')

my_dict['occupation'] = 'ML Engineer' # Add new key-value pair
my_dict['age'] = 22               # Update existing key's value

print(f'Dictionary after adding occupation and updating age: {my_dict}')
```

```python
print('\n4. Removing Elements: Using del, .pop(), and .popitem().')

# Using del statement
del my_dict['city']
print(f'Dictionary after deleting city with del: {my_dict}')

# Using .pop() (removes specified key and returns its value)
removed_occupation = my_dict.pop('occupation')
print(f'Removed occupation: {removed_occupation}')
print(f'Dictionary after popping occupation: {my_dict}')

# Using .popitem() (removes and returns the last inserted key-value pair)
# In Python 3.7+, insertion order is preserved.
last_item = my_dict.popitem()
print(f'Removed last item: {last_item}')
print(f'Dictionary after popping last item: {my_dict}')
```

```python
print('\n5. Dictionary Methods: .keys(), .values(), .items().')

my_dict = {
    'fruit': 'apple',
    'color': 'red',
    'quantity': 5
}

print(f'Dictionary: {my_dict}')
print(f'Keys: {my_dict.keys()}')
print(f'Values: {my_dict.values()}')
print(f'Items (key-value pairs): {my_dict.items()}')
```

```python
print('\n6. Looping Through Dictionaries: Iterating over keys, values, or items.')

print('Looping through keys:')
for key in my_dict:
    print(f'  Key: {key}')

print('\nLooping through values:')
for value in my_dict.values():
    print(f'  Value: {value}')

print('\nLooping through items:')
for key, value in my_dict.items():
    print(f'  {key}: {value}')
```

```python
print('\n7. Copying Dictionaries: Shallow copy with .copy().')

dict_original = {'a': 1, 'b': 2}
dict_copy = dict_original.copy() # Creates a shallow copy

dict_copy['c'] = 3

print(f'Original dictionary: {dict_original}')
print(f'Copied dictionary (with new element): {dict_copy}')

# Important: For nested dictionaries, .copy() creates a shallow copy.
# Changes to nested objects in the copy will affect the original.
# For deep copy, use copy.deepcopy() from the 'copy' module.
import copy
dict_nested = {'data': [1, 2, 3]}
dict_deep_copy = copy.deepcopy(dict_nested)
dict_deep_copy['data'].append(4)

print(f'Nested original: {dict_nested}')
print(f'Nested deep copy (with appended element): {dict_deep_copy}')
```

```python
print('\n8. Updating dictionary elements : ')

dict_1 = {1 : 'Kushagra' , 2 : 'Ayush' , 3 : 'Kirti'}
dict_1[4] = 'Shardul'   # I added a new key-value
print(f"Manually adding a new Key and it's value : {dict_1}")

dict_1.update({5 : 'Yashi' , 6 : 'Vivek'})   # using .update() function I can update / add multiple key-values pairs in the existing dictionary

print(f"Printing the updated dictionary : {dict_1}")
```

```python
print('\n9. Iteration of key , value in dictionary : \n')

dict_a = {1 : 'apple' , 2 : 'mango' , 3 : 'litchi'}

print("Printing key , value of dictionary : ")
for key , value in dict_a.items():
  print(key , value)

print("")

print("Printing keys of dictionary : ")
for key in dict_a.keys():
  print(key)

print("")

print("Printing values of dictionary : ")
for value in dict_a.values():
  print(value)
```

# ***Frequency of Elements***

```python
# first I will count the freq of elements in a list

l = ["Kush" , "Kirti" , "Yash" , "Aman" , "Kush" , "Yashi" , "Vivek" , "Kirti" , "Yash" , "Yash" , "Vivek"]

hash_map = {}   # I will make an empty dictionary

for name in l:
  if name not in hash_map:
    hash_map[name] = 1
  else:
    hash_map[name] = hash_map[name] + 1
print(hash_map)
```

```python
# similar example for counting characters of a string

s = "Kushagra has gr8 ML skills"

freq_map = {}

for char in s:
  if char not in freq_map:
    freq_map[char] = 1
  else:
    freq_map[char] = freq_map[char] + 1

print(freq_map)
```

# ***Hashing*** -
it is a technique used in daat structures that efficiently stores and retrieves data in a way that allows for quick access.

```python
from typing import List

class Solution:
  def twoSum(self , nums : List[int] , target : int) -> List[int]:
    n = len(nums)
    hash_map = {}

    for i in range(n):
      diff = target - nums[i]       # I calculate the difference of target and the i-th number
      if diff == 0:
        return [i]

      elif diff < 0:
        return None

      elif diff in hash_map:          # what I do is , is difference is present in hash_map then i return its index along with the index - i , becoz i-th number and the number in dictionary(hash_map) sums equal to target
        return [hash_map[diff] , i]

      else:
        hash_map[nums[i]] = i    # otherwise if the difference isn't present , in normal case the difference gets stored in hash_map

s = Solution()
```

```python
print(s.twoSum([2 , 7 , 11 , 15] , target = 9))
```

Explanation - we are on a number at i-th location , we neeed a target , so we storw the difference btw number and that target in a hash_map. if next number while iterating is equal to the difference stored in a hash_map , I return its value along with the current i-th index.
