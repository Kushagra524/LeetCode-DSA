# ***Kadane's Algorithm***

#### *This algo. is used to find maximum sum subarray. For example : arr = [5 , -1 , -7 , 6 , 2 , -3 , 5 , -10 , 2] . Now in this array , I want to find which subarray has maximum sum. So output = [6 , 2 , -3 , 5] and subarray can be of any size but ofcourse len(subarray) <= len(array)*

So how this algo. works:

CASE 1 : all +ve elements ==> arr1 = [1 , 2 , 3 , 4]   # , then output = sum of all elements = 10


CASE 2 : all -ve elements ==> arr2 = [-1 , -2 , -3 , -4]  # greatest element = output = -1

So , we take two var's : current_sum = 0 and max_sum = nums[0] becoz initially max sum is first element only.

# ***Example***

image.png

```python
from typing import List

class Solution:
  def maxSubArray(self , nums : List[int]) -> int:
    curr_sum = 0
    max_sum = nums[0]    # I am taking max_sum = 1st element

    for i in range(0 , len(nums)):
      curr_sum = curr_sum + nums[i]
      if curr_sum > max_sum:
        max_sum = curr_sum
      if curr_sum < 0:
        curr_sum = 0
    return max_sum

s = Solution()
```

*Explanation : first I iterate in my array , and current = curr_sum + i-th element. then I check if current_sum > max_sum , then my max_sum becomes equal to current_sum obviously becoz it's value increased. And if current_sum < 0 (means -ve sums) then we make current sum = 0 becoz if I want maximum sum subarray then why should I consider -ve subarray sum becoz we see that their are +ve elements too , and start fresh calculation of max and curr sum again till I reach last element of array*

```python
print(s.maxSubArray([-2,1,-3,4,-1,2,1,-5,4]))
```
