# ***Pow(x , n)***

***Recursive approach - O(log n)***

```python
class Solution:
  def power(self, x , n):
    # first i will make the base case
    if n == 0:
      return 1

    # recursive case
    a = self.power(x , n // 2)
    if n % 2 == 0:
      return a*a
    elif n % 2 != 0:
      return a * a * x

  def myPow(self , x : float , n : int) -> float:
    if n >= 0:
      return self.power(x , n)
    else:
      return 1 / self.power(x , n * (-1))

s = Solution()
```

```python
print(s.myPow(2 , 10))
```

***Time complexity : O(log n)***

# ***Explanation of above code :***

## first what I did is , base case if n = 0 , means x ^ 0 = 1 so return 1

## then in recursive case , each and everytime , power become half ( n // 2) and stores in a variable a , so a = x ^ (n // 2) now : here 2 cases arises - if power is even , then a * a will be executed , otherwise in case of odd power : a * a * x

## this functions repeats till n = 0 : and eventually we get x ^ n as ouput

***Another method - O(n)***

```python
class Solutions:
  def myPow(self , x , n ):
    result = 1

    while n > 0:
      result = result * x
      n = n - 1

    return result

s1 = Solutions()
```

```python
print(s1.myPow(3 , 5))
print(s1.myPow(4 , 3))
```
