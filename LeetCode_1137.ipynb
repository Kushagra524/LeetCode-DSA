# ***N-th Tribonacci Series***

image.png

```python
class Solution:
  def tribonacci(self , n : int) -> int:
    if n == 0:
      return 0
    elif n == 1 or n == 2:
      return 1
    a = 0
    b = 1
    c = 1

    for i in range(3 , n+1):
      result = a + b + c
      a = b
      b = c
      c = result
    return c

s = Solution()

print(s.tribonacci(0))
print(s.tribonacci(1))
print(s.tribonacci(2))
print(s.tribonacci(3))
print(s.tribonacci(4))
```
