```python
from typing import List

class Solution:
  def twoSum(self , nums : List[int] , target : int) -> List[int]:
    n = len(nums)
    hash_map = {}

    for i in range(n):
      diff = target - nums[i]

      if diff == 0:
        return [i]

      elif diff < 0:
        return None

      elif diff in hash_map:
        return [hash_map[diff] , i]

      else:
        hash_map[nums[i]] = i

s = Solution()
```

```python
print(s.twoSum([2 , 7 , 11 , 15] , target = 9))
```

Explanation - The technique used here is called hashing.

Initially i create an empty hash_map

Then we iterate through nums , and calculate the difference , at each i-th index , we calculate the difference ( target - nums[i] )

Then if diff present in hash_map , we return a list =  [hash_map[diff] , i] where hash+mpa[diff] is the index of that difference in hash_map , and 'i' is the index of element present in nums ===> Ofcoure hash_maps[diff] + nums[i] = target.

Else , that element is added into hash_map with the particular frequency.

```python
from typing import List

class Solution:
  def twoSum(self , nums : List[int] , target : int) -> List[int]:

    hash_map = {}

    for i in range(len(nums)):
      diff = target - nums[i]
      if diff == 0:
        return [i]

      elif diff in hash_map:
        return [hash_map[diff] , i]

      else:
        hash_map[nums[i]] = i

s = Solution()
```
