# ***Intro. To Recursion :***

*The process in which a function calls itself directly or indirectly until it reached the base case is called recursion.*
*And the corresponding function as called recursive function*

```python
# comparision btw loop and recursion using an example

# printing i to n using loops
n = 5
i = 1

while i <=n:
  print(i , end = ' ')
  i = i + 1

  # this is an example of loop
```

```python
# same above example using recursion
def printNums(i , n):
  # base case
  if i > n:
    return
  # recursive case
  print(i , end = ' ')
  # here i am calling my function recursively
  printNums(i+1 , n)

printNums(1 , 5)
```

```python
# example - 2

# factorial program using recursion

def factorial(n):
  if n == 1 or 0:
    return 1
  return n * factorial(n - 1)

factorial(5)
```

***Understanding Recursive Stack***

```python
def fun(n):
  if n == 0:
    return
  fun(n - 1)
  print(n , end = ' ')

fun(3)

# what happens here is first fun(3) called ---> goes into function ---> checks if base case is present ---> no --> then fun(3 - 1) called recursively --> then again fun(3 - 1 - 1) called ----> still base case isn't achieved yet --> fun(3 - 1 - 1 - 1) means fun(0) called = base case says return mean end of funtion = goes to previous funtion prints fun(1) = prints 1 --> then end of this function --> fun(2) = prints 1  2 --> goes to fun(3) = prints 1 2 3 ---> output comes = 1 2 3
```
