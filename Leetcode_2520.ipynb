# ***2520. Count the Digits That Divide a Number***

Screenshot 2026-06-25 173210.png

```python
class Solution:
    def countDigits(self, num: int) -> int:
        temp = num
        result = []
        while temp > 0:
            id = temp % 10
            result.append(id)
            temp = temp // 10
        count = 0
        for i in result:
            if num % i == 0:
                count += 1
        return count
```

```python
s = Solution()
```

```python
s.countDigits(121)
```

```python
s.countDigits(1248)
```
