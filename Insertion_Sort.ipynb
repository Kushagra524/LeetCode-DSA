# ***Insertion Sort*** -


```python
from typing import List

class Solution:
  def insertionSort(self , arr : List[int]) -> List[int]:
    n = len(arr)

    for i in range(1 , n):
      key = arr[i]
      j = i - 1

      while j >= 0 and key < arr[j]:
        arr[j + 1] = arr[j]
        j = j - 1

      arr[j + 1] = key
    return arr

s = Solution()
```

```python
print(s.insertionSort([40 , 20 , 30 , 10]))
```

***For desceding order***

```python
from typing import List

class Solution:
  def insertionSort(self, arr: List[int]) -> List[int]:
    n = len(arr)

    for i in range(1, n):
      key = arr[i]
      j = i - 1

      while j >= 0 and arr[j] < key:    # for descending order
        arr[j + 1] = arr[j]
        j -= 1

      arr[j + 1] = key

    return arr


s = Solution()
```

```python
print(s.insertionSort([40, 20, 30, 10]))
```
