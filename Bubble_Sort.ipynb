# ***Bubble Sort*** - [5 , 3 , 4 , 1 , 2]
Pass 1. ===> I take 1st element & comapre it from 2nd element , if 1st element greater = swapped if not we move on 2nd index and same logic goes on. In pass 1 , 1 element is set at it's correct position. [3 , 4 , 1 , 2 , 5]

Pass 2. Again same  logic starting from 1st element and comparison with 2nd element and comparison , if true = Swap.
[3 , 1 , 2 , 4 , 5]

Pass 3. Then in pass 3 array becomes [1 , 2 , 3 , 4 , 5].

Pass 4. Now we got sorted array but still we check each and every element to get confirmation that array is sorted. Becoz for n elements , we make (n - 1) passes.

Outer loop - repeats sorting process and controls number of passes , that's why it runs (n - 1) times.

Inner loop - performs sorting of one element at a time by comparision with it's element ahead that's why it runs (n - i - 1) times.

```python
from typing import List

class Solution:
  def bubbleSort(self , arr : List[int]) -> List[int]:
    n = len(arr)

    for i in range(0 , n - 1):
      for j in range(0 , n - i -1):
        if arr[j] > arr[j + 1]:            # for ascending order
          arr[j] , arr[j + 1] = arr[j + 1] , arr[j]


    return arr

s = Solution()
```

```python
print(f"Ascending order using Bubble Sort : {s.bubbleSort(arr = [5 , 3 , 8 , 4 , 2])}")
```
