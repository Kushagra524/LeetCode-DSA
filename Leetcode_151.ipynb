# ***151. Reverse Words in a String***

image.png

```python
class Solution:
  def reverseWords(self , s : str) -> str:
    s = s.strip()      # .strip() removes the endin spaces at both ends
    s = s.split()      # .split() is used to convert the string elements to list

    left = 0
    right = len(s) - 1
    while left < right:
      s[left] , s[right] = s[right] , s[left]

      left = left + 1
      right = right - 1
    return " ".join(s)

s = Solution()
```

```python
print(s.reverseWords("        I am      an    ML Developer     "))
```

*Another Approach without using in-built function for removing spaces*

```python
class Solutions:
  def reversedWords(self , s : str) -> str:
    s = s.split()
    for i in s:
      if i == " ":
        s.remove(i)

    left = 0
    right = len(s) - 1

    while left < right:
      s[left] , s[right] = s[right] , s[left]

      left = left + 1
      right = right - 1
    return " ".join(s)

s1 = Solutions()
```

```python
print(s1.reversedWords("        I am      an    ML Developer     "))
```
